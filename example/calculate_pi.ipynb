> 半径1の円に内接する正$n$角形の面積

半径1の円に内接する正$n$角形の面積を$S_n$とすると

$$
S_n = \frac{n}{2} \times \sin \frac{2\pi}{n}
$$

半径1の円の面積は$\pi$なので以下のことが予想される

$$
\lim_{n\to\infty} S_n = \pi
$$





```
(*Short Answerの確認*)
WolframAlpha["Limit[(x/2)sin[2Pi/x], x ->\[Infinity]", "PodCells"]
```
